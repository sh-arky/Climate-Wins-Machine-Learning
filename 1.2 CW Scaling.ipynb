## 1. Importing Libraries and Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import os
import sklearn
from sklearn.preprocessing import StandardScaler
```

```python
path = r"C:\Users\Noah\ClimateWins"
```

```python
df = pd.read_csv(os.path.join(path, '02 Data', 'Raw', 'Dataset-weather-prediction-dataset-processed.csv'))
```

## 2. Cleaning

```python
# indexing data points with individual numbers assigned
df.reset_index(inplace=True)
df.rename(columns={'index':'id'}, inplace=True)
df.head()
```

```python
df.info()
```

```python
# converting DATE column into dates
df['DATE'] = pd.to_datetime(df['DATE'].astype(str), format='%Y%m%d')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

#### no null, can continue to scaling

## 3. Scaling

```python
# initialize the standard scaler
scaler = StandardScaler()
```

```python
#Assuming all columns except 'DATE' need to be scaled
scales = df.columns[df.columns != 'DATE']  # can adjust if some columns should not be scaled
df[scales] = scaler.fit_transform(df[scales])
```

```python
df.head()
```

## 4 Exporting

```python
# Save the scaled data back to CSV
df.to_csv(os.path.join(path, '02 Data', 'Cleaned', '[DATASET]_scaled.csv'))
```
